```python
# Load the data
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("dataset/properties3.csv")
# print(df.head()) # show the table

# using a histogram
plt.hist(df["price"], bins=50)
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()

# boxplot
sns.boxplot(x="district", y="price", data=df)
plt.show()

## line plot of prices over time
df["date"] = pd.to_datetime(df["date"].apply(lambda x: "01-" + x), format="%d-%b-%y")
df = df.sort_values(by="date")
prices_by_date = df.groupby("date")["price"].mean()

plt.plot(prices_by_date.index, prices_by_date.values)
plt.xlabel("Date")
plt.ylabel("Price")
plt.show()


# Scatterplot of Prices vs Area
plt.scatter(df["area"], df["price"])
plt.xlabel("Area")
plt.ylabel("Price")
plt.show()
```
