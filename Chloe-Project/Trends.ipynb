```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the xlsx file
filename = "personnel.xlsx"
df = pd.read_excel(filename, engine='openpyxl', index_col='Data Series')

# Remove the 'TOTAL' row from the DataFrame
df = df.iloc[:-1]

# Remove rows with 'NaN' values
df_cleaned = df.dropna()

# Transpose the DataFrame to have years as rows and roles as columns
df_transposed = df_cleaned.transpose()

# Define a list of six shades of blue
blue_colors = ['#cfe6f8', '#9fcde6', '#6da7d9', '#4781c1', '#2c5f96', '#1d3c6c']


# Plot the data with different shades of blue
fig, ax = plt.subplots(figsize=(10, 6))
df_transposed.plot(kind='bar', ax=ax, color=blue_colors)
ax.set_title("Healthcare Workers Distribution (2018-2021)")
ax.set_xlabel("Year")
ax.set_ylabel("Number of Workers")
ax.legend(title="Roles", loc='upper left', bbox_to_anchor=(1, 1))

# Save the figure
plt.savefig("healthcare_workers_distribution.png", bbox_inches='tight')

# Show the plot in Jupyter Notebook
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the xlsx file and the specific sheet
filename = "bed-occupancy-rate.xlsx"
sheet_name = "BOR(%)_historical"
df = pd.read_excel(filename, engine='openpyxl', sheet_name=sheet_name, header=2)

# Convert 'Date' column to datetime objects
df['Date'] = pd.to_datetime(df['Date'])

# Drop the 'Years' column and the 'SKH' column with NaN values
df = df.drop(columns=['Years', 'SKH'])

# Set 'Date' column as the index
df = df.set_index('Date')

# Resample data by week, month, and year
df_weekly = df.resample('W').mean()
df_monthly = df.resample('M').mean()
df_yearly = df.resample('Y').mean()

# Plot the data
def plot_data(df, title, filename):
    fig, ax = plt.subplots(figsize=(10, 6))
    df.plot(kind='line', marker='o', ax=ax)
    ax.set_title(title)
    ax.set_xlabel("Date")
    ax.set_ylabel("Bed Occupancy Rate (%)")
    ax.legend(title="Hospitals", loc='upper left', bbox_to_anchor=(1, 1))
    
    # Save the figure
    plt.savefig(filename, bbox_inches='tight')

    # Show the plot in Jupyter Notebook
    plt.show()

plot_data(df_weekly, "Weekly Bed Occupancy Rate (BOR) for Hospitals (2018-2023)", "bed_occupancy_rate_weekly.png")
plot_data(df_monthly, "Monthly Bed Occupancy Rate (BOR) for Hospitals (2018-2023)", "bed_occupancy_rate_monthly.png")
plot_data(df_yearly, "Yearly Bed Occupancy Rate (BOR) for Hospitals (2018-2023)", "bed_occupancy_rate_yearly.png")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the xlsx file and the specific sheet
filename = "bed-occupancy-rate.xlsx"
sheet_name = "BOR(%)_historical"
df = pd.read_excel(filename, engine='openpyxl', sheet_name=sheet_name, header=2)

# Convert 'Date' column to datetime objects
df['Date'] = pd.to_datetime(df['Date'])

# Drop the 'Years' column and the 'SKH' column with NaN values
df = df.drop(columns=['Years', 'SKH'])

# Set 'Date' column as the index
df = df.set_index('Date')


# Custom colors (shades of blue)
colors = [
    "#1F77B4",
    "#2E9BCA",
    "#40B6E0",
    "#61C8EF",
    "#89D7F3",
    "#B0E6F7",
    "#D4F2FB",
]

df_monthly = df.resample('M').mean()
df_yearly = df.resample('Y').mean()

# Filter and plot data for specific years
def plot_data_by_years(df, start_year, end_year, title, filename):
    df_filtered = df.loc[start_year:end_year]
    fig, ax = plt.subplots(figsize=(10, 6))
    df_filtered.plot(kind='line', marker='o', ax=ax, color=colors)
    ax.set_title(title)
    ax.set_xlabel("Date")
    ax.set_ylabel("Bed Occupancy Rate (%)")
    ax.legend(title="Hospitals", loc='upper left', bbox_to_anchor=(1, 1))

    # Save the figure
    plt.savefig(filename, bbox_inches='tight')

    # Show the plot in Jupyter Notebook
    plt.show()

plot_data_by_years(df_monthly, '2018', '2019', "Bed Occupancy Rate (BOR) for Hospitals (2018-2019)", "bed_occupancy_rate_2018_2019.png")
plot_data_by_years(df_monthly, '2020', '2021', "Bed Occupancy Rate (BOR) for Hospitals (2020-2021)", "bed_occupancy_rate_2020_2021.png")
plot_data_by_years(df_monthly, '2022', '2023', "Bed Occupancy Rate (BOR) for Hospitals (2022-2023)", "bed_occupancy_rate_2022_2023.png")

plot_data_by_years(df_yearly, '2018', '2019', "Bed Occupancy Rate (BOR) for Hospitals (2018-2019)", "bed_occupancy_rate_2018_2019.png")
plot_data_by_years(df_yearly, '2020', '2021', "Bed Occupancy Rate (BOR) for Hospitals (2020-2021)", "bed_occupancy_rate_2020_2021.png")
plot_data_by_years(df_yearly, '2022', '2023', "Bed Occupancy Rate (BOR) for Hospitals (2022-2023)", "bed_occupancy_rate_2022_2023.png")

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# import data
file_path = "4. M870041.xlsx"
df = pd.read_excel(file_path, sheet_name="cleaned")

# Set the Data Series column as the index and drop the public sector hospital admissions row
df.set_index("Data Series", inplace=True)
df.drop("Public Sector Hospital Admissions", axis=0, inplace=True)

# Transpose
df = df.T

# Convert the index to datetime objects
df.index = pd.to_datetime(df.index, format="%Y %b ")

df.drop("Total", axis=1, inplace=True)

# need to have blue
colors = sns.color_palette("Blues", n_colors=len(df.columns))

# Plot the line graph
plt.figure(figsize=(12, 6))
for i, column in enumerate(df.columns):
    plt.plot(df.index, df[column], label=column, color=colors[i])

plt.xlabel("Month")
plt.ylabel("Hospital Admissions")
plt.title("Hospital Admissions Over Time")
plt.legend()
plt.show()

plt.figure(figsize=(12, 6))
sns.boxplot(data=df, palette="Blues")
plt.xlabel("Hospital")
plt.ylabel("Hospital Admissions")
plt.title("Hospital Admissions Distribution")
plt.xticks(rotation=90)
plt.show()
```
